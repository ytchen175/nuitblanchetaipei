```python
import requests 
from lxml import etree
import re
# url = 'https://www.nuitblanchetaipei.info/artwork'
url = 'https://www.nuitblanchetaipei.info/artwork'
headers = {
    'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/604.5.6 (KHTML, like Gecko) Version/11.0.3 Safari/604.5.6'
    }
res = requests.get(url,headers=headers)
print(res.text)
```

```python
xdoc = etree.HTML(res.text)
# act_name = map(str.strip , xdoc.xpath("//tr[2]//div[@class='box-heading']/h2/span[@class='ch']/text()"))
act_name = xdoc.xpath("//tr[2]//div[@class='box-heading']/h2/span[@class='ch']/text()")
artist = map(str.strip , xdoc.xpath("//tr[2]//div[@class='box-heading']/span[@class='artist']/text()"))
# description = map(str.strip , xdoc.xpath("//tr[2]//div[@class='artwork_box box m-1 text-left']/a[@class='box-anchor']/p/text()"))
# time = map(str.strip , xdoc.xpath("//tr[2]//div[@class='artwork_box box m-1 text-left']/dl[@class='details']/dd[1]/text()"))
time = xdoc.xpath("//tr[2]//div[@class='artwork_box box m-1 text-left']/dl[@class='details']/dd[1]/text()")
location = map(str.strip , xdoc.xpath("//tr[2]//div[@class='artwork_box box m-1 text-left']/dl[@class='details']/dd[2]/text()"))
```

```python
# 如果遇到\r \n則將list的join在一起，replace \r\t to space and join
```
